## Загрузка библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Создание данных для построения графиков

```python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['a']) * 1.5
data['f'] = data['a'] + 5 * np.random.randn(50)
data['g'] = data['d'] + data['a']
data['h'] = data['g'] // 2
data['j'] = data['f'] + 10
data['k'] = data['j'] + data['a'] * 2
data['m'] = data['c'] // 2
data['n'] = data['m'] + 15
data['p'] = data['g'] // 2 + 20

df = pd.DataFrame(data)
df.head()
```

## Примеры построения графиков

### С помощью метода класса pandas.DataFrame

```python
df.plot(x='a',
        y='c',
#         kind='scatter', # Допускается задавать вид графика. При этом меняются его гиперпараметры настройки.
        figsize=(5, 4), # Определение размера изображения
        title='График', # Определение названия графика
        xlabel='Шкала Х', # Определение названия шкалы X
        ylabel='Шкала Y', # Определение названия шкалы Y
        xlim=(0, 50), # Определение максимального и минимального значений на шкале X
        ylim=(-10, 60), # Определение максимального и минимального значений на шкале Y
        xticks=(np.arange(0, 60, 5)), # Определение рисок на шкале X
        yticks=(np.arange(-10, 60, 5)), # Определение рисок на шкале Y
        grid=True, # Включение сетки на графике
        marker='o', # Определение вида маркера (например '.', ',', 'v', '^', '<', '>', 'x' и др.)
        markersize=5, # Определение размера маркера
        linestyle='-', # Определение вида линии
        linewidth=1, # Определение толщины линии
        color='C3' # Определение цвета линии (от C0 до C9)
       )
```

### Построение простого графика с библиотекой matplotlib

```python
# Примеры построения зависимостей различных данных
# от значений колонки "а" разными линиями
plt.plot(df['a'], df['b'], linewidth=1, label='1st component')
plt.plot(df['a'], df['d'], ':b', linewidth=2, label='2nd component') # пунктирная синяя линия
plt.plot(df['a'], df['c'], '--r', linewidth=3) # штрихованная красная линия
plt.plot(df['a'], df['f'], 'ob')  # синие круги
plt.plot(df['a'], df['g'], 'vr')  # красные треугольники
plt.plot(df['a'], df['h'], 'Dk')  # черные ромбы
plt.plot(df['a'], df['p'], linewidth=2, marker='o', c='C6')

# Добавление надписей
plt.title('График') # Добавление названия графика
plt.legend(fontsize=7) # Добавление легенды
plt.xlabel("Значения X", fontsize=16) # Добавление названия оси X + настройка размера шрифта
plt.ylabel("Значения Y") # Добавление названия оси Y

# Настройка шкал
plt.xlim(0, 50)  # Диапазон оси X от 0 до 50
plt.ylim(-10, 80)  # Диапазон оси Y от -10 до 80
plt.xticks(np.arange(0, 60, 10), fontsize=5) # Изменяем шаг делений на оси X, вводятся значения меток на оси
plt.yticks(np.arange(-10, 61, 10)) # Изменяем шаг делений на оси Y, вводятся значения меток на оси

# Настройка сетки
plt.grid(which='major') # Выводим основную сетку
plt.minorticks_on() # Включаем вспомогательные линии сетки
plt.grid(which='minor', linestyle=':') # Выводим вспомогательную сетку

plt.show()
```

### Построение сетки графиков с библиотекой matplotlib

```python
cols = df.columns[1:] # первый элемент исключен, так как все графики будут строится от него

fig, axes = plt.subplots(ncols=3, nrows=2, figsize=(12, 8)) # Задаём область графиков
fig.suptitle('График зависимостей параметров от а', fontsize=20) # Выводим название всей области графиков

for idx, row in enumerate(axes):
    for jdx, ax in enumerate(row):
        ndf = idx * 3 + jdx + 1 # Определение индекса в списке cols
        ax.plot(df['a'], df[cols[ndf]], label=f'Зависимость {cols[ndf]} от а') # Строим линию зависимости
        
        ax.set_title(f'График зависимости {cols[ndf]} от а') # Добавление названия графика
        ax.set_xlabel('Параметр а', fontsize=16) # Добавление названия оси X + настройка размера шрифта
        ax.set_ylabel(f'Параметр {cols[ndf]}', fontsize=16) # Добавление названия оси Y + настройка размера шрифта
        ax.legend(fontsize=10) # Добавление легенду
        
        ax.set_xlim([0, 50]) # Диапазон оси X от 0 до 50
        ax.set_ylim([-10, 80]) # Диапазон оси Y от -10 до 80
        ax.set_xticks(np.arange(0, 55, 5), fontsize=5) # изменяем шаг делений на оси X, вводятся значения меток на оси
        ax.set_yticks(np.arange(-10, 91, 10)) # изменяем шаг делений на оси Y, вводятся значения меток на оси
        
        ax.grid(which='major') # Выводим основную сетку
        ax.minorticks_on() # Включаем вспомогательные линии сетки
        ax.grid(which='minor', linestyle=':') # Выводим вспомогательную сетку

fig.tight_layout()
plt.show()
```

Более тонкую настройку параметров можно выполнить, передавая различные именованные аргументы, например:

* marker: str — тип маркера
* markersize: float — размер маркера
* linestyle: str — тип линии
* linewidth: float — толщина линии
* color: str — цвет
